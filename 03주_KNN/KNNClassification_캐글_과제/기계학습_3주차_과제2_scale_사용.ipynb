# KNN회귀를 이용한 배추값 예측 

```
import numpy as np
import pandas as pd
```

#### 데이터 불러오기

```
train_data = pd.read_csv('/content/train_cabbage_price.csv')
test_data = pd.read_csv('/content/test_cabbage_price.csv')
print(train_data)
print(type(train_data))
print(train_data.shape)
print(train_data.info())
```

- 결측치 없음
- 클래스 열 자료형 int형

```
print(test_data)
print(type(test_data))
print(test_data.shape)
print(test_data.info())
```

#### train 데이터 분리

```
train_data_np = np.array(train_data)
x_train_np = train_data_np[:,:5]
y_train_np = train_data_np[:,5]

print(x_train_np)
print(type(x_train_np))
print(x_train_np.shape)

print(y_train_np)
print(type(y_train_np))
print(y_train_np.shape)

x_train_df = pd.DataFrame(x_train_np)
y_train_df = pd.DataFrame(y_train_np)
print(x_train_df)
print(type(x_train_df))
print(x_train_df.shape)

print(y_train_df)
print(type(y_train_df))
print(y_train_df.shape)

```

## 데이터 변환(minmaxscaler)
: 데이터를 0-1 사입 값으로 변환

#### train 데이터 scale

```
 from sklearn.preprocessing import MinMaxScaler
 scaler = MinMaxScaler()
 scaler.fit(x_train_df)
 x_train_df_scale = scaler.transform(x_train_df)
 print(x_train_df_scale)
 print(type(x_train_df_scale))
 print(x_train_df_scale.shape)



 #print(scaler.transform([[2, 2]]))
```

#### test 데이터 scale

```
 from sklearn.preprocessing import MinMaxScaler
 scaler = MinMaxScaler()
 scaler.fit(test_data)
 x_test_df_scale = scaler.transform(test_data)
 print(x_test_df_scale)
 print(type(x_test_df_scale))
 print(x_test_df_scale.shape)


```

#### 스케일 결과
: 이 데이터 에서는 모델의 성능이 오히려 떨어져서 쓰지 않는 것이 좋다.

## 모델 설계 및 학습

#### knn 회귀

```
from sklearn.neighbors import KNeighborsRegressor
```

```
knnReg = KNeighborsRegressor(n_neighbors=3,p=2,weights='distance')
knnReg.fit(x_train_df_scale,y_train_df)
```

#### 예측값 

```
y_test_pred = knnReg.predict(x_test_df_scale)
print(y_test_pred)
print(type(y_test_pred))
print(y_test_pred.shape)
```
