## 데이터 cleaning


### 결측치 처리
- 결측치 제거
- 결측치 채우기(평균,중앙,최빈)

```
import numpy as np

# 결측치 처리
x_miss = np.array([[1,2,3,None],[5,np.nan,7,8],[None,10,11,12],[13,np.nan,15,16]])
x_miss
```

```
from sklearn.impute import SimpleImputer # 교수님과 버전 달라
im = SimpleImputer(strategy='mean')
x_full = im.fit_transform(x_miss) 
print(x_full)
print(x_miss)
```
