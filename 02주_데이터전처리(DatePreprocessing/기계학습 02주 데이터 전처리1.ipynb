### 범주형 자료의 실수화
: 원핫 인코딩

```
# 범주형 자료형
x=[{'city':'seoul','temp':10.0},{'city':'Dubai','temp':23.0},{'city':'LA','temp':20.0}]
x
```

```
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer(sparse=False) # sparse=True 는 불필요한거 없애서 데이터 확보
vec.fit_transform(x)
```

### 텍스트 자료의 실수화

```
text =['떴다 떴다 비행기 날아라 날아라',
       '높이 높이 날아라 우리 비행기',
       '내가 만든 비행기 날아라 날아라',
       '멀리 멀리 날아라 우리 비핼기']
```

```
from sklearn.feature_extraction.text import CountVectorizer
vec2 = CountVectorizer() # 디폴트값이 sparse=True 
t = vec2.fit_transform(text).toarray() # sparse=false로 하고 text를 수 배열로 변환
import pandas as pd
t1 = pd.DataFrame(t,columns = vec2.get_feature_names()) # 각 열의 항목에 이름들을 가져온다
t1
```

#### tf-idf
: 의미 없는 단어의 가중치를 줄이고자 비중많은 것은 가중치 줄이고 적은것은 높임

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
x2 = tfid.fit_transform(text).toarray()
x3 = pd.DataFrame(x2,columns = tfid.get_feature_names())
x3
```
