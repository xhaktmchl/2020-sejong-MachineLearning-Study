## 데이터 불러오기

```
import seaborn as sns
```

```
iris = sns.load_dataset('iris')

print(iris)
print(type(iris))
print(iris.shape)
print(iris.info())
```

#### x,y 데이터 나누기

```
import pandas as pd
```

```
X_data = iris.iloc[:,:4]
y_data = iris.iloc[:,4]

print(X_data)
print(type(X_data))
print(X_data.shape)


print(y_data)
print(type(y_data))
print(y_data.shape)
```

#### 클래스 라벨 인코딩

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y_data)

le.classes_

y_data = le.transform(y_data)

print(y_data)
print(type(y_data))
print(y_data.shape)
```

#### train test 셋 나누기

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(
    X_data, y_data, test_size=0.33, random_state=42)

print(x_train)
print(type(x_train))
print(x_train.shape)
print(x_test)
print(type(x_test))
print(x_test.shape)
print(y_train)
print(type(y_train))
print(y_train.shape)
print(y_test)
print(type(y_test))
print(y_test.shape)
```

## 로지스틱회귀 실행

LogisticRegression(C=200, random_state=0).fit(x_train,y_train)
- C= 양의 실수값  : 정규화의 정도를 정하는 속성

```
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0).fit(x_train,y_train)

y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)

print(y_train_pred)
print(type(y_train_pred))
print(y_train_pred.shape)
print(y_test_pred)
print(type(y_test_pred))
print(y_test_pred.shape)

# 여기부터는 함수 뭔지 알려고 실행 해봄
print(clf.predict_proba(x_train))


print(clf.score(x_train, y_train))
```

#### 모델 정확도 평가

```
from sklearn.metrics import accuracy_score

print(accuracy_score(y_train, y_train_pred, normalize=False)) # 정규화 않한 것
print(accuracy_score(y_test, y_test_pred)) # 정규화 디폴트 값

```
